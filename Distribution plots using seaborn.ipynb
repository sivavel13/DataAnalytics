# Seaborn

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Import Data

```python
# Seaborn provides built in datasets
print(sns.get_dataset_names())
```

```python
# Load a built in dataset based on US State car crash percentages
tips = sns.load_dataset("tips")
tips.head(10)
```

## Distribution Plots

### Distributin plot

```python
# kde means convert count to percentage values
sns.distplot(tips["total_bill"],kde = True, bins = 25)
plt.title("Distribution of Total bill amount")
```

```python
# When pass KDE = False, It's act like Histogram because y axis mention actual count
sns.distplot(tips["total_bill"], kde = False, bins = 25)
```

## Joint Plot

Jointplot allows you two basically match up two distplot for bivariant data. With your choice of what kind of parameter to compare with:
1) Scatter 
2) reg
3) resid
4) kid
5) hex

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter",hue="sex")
plt.title("Distibution of tips and total bills using scatter plot", pad=90)
```

```python
# Can change kind paramas give different values for different plots
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
plt.title("Distibution of tips and total bills using regression plot", pad=90)
```

## Pair plot

```python
sns.pairplot(tips, hue = "sex", palette = "coolwarm")
plt.title("Correlation for numeric values from dataframe", pad=350, loc="right")
```

## Rugplot

```python
sns.rugplot(tips["total_bill"])
sns.kdeplot(tips["total_bill"])
```

```python
sns.rugplot(tips["tip"])
sns.kdeplot(tips["tip"])
plt.title("Density of tips")
```
